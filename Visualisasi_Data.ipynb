```python
import pandas as pd
import matplotlib.pyplot as plt

# Membaca Dataset
df = pd.read_csv("student_performance_encoded.csv")

# Menampilkan Data
print("="*60)
print("10 BARIS PERTAMA DATASET")
print("="*60)
print(df.head(10))

print("\nJumlah Baris dan Kolom")
print(df.shape)

print("\nInformasi Dataset")
print(df.info())

print("\nStatistik Deskriptif")
print(df.describe())

# Data untuk Visualisasi
cgpa = df["cgpa"].value_counts().sort_index()
academic = df["Academic_Satisfaction"].value_counts().sort_index()

# Label kategori
cgpa_labels = ["5.0–6.9", "7.0–8.4", "8.5–9.4", "9.5–10.0"]

academic_labels = [
    "Very\nUnsatisfied",
    "Unsatisfied",
    "Neutral",
    "Satisfied",
    "Very\nSatisfied"
]

# Membuat 2 Subplot
fig, ax = plt.subplots(1, 2, figsize=(16,6))

# SUBPLOT 1 : Grafik Batang CGPA
bars = ax[0].bar(
    cgpa_labels,
    cgpa.values,
    color="skyblue",
    edgecolor="black",
    label="Jumlah Mahasiswa"
)

ax[0].set_title("Distribusi Kategori CGPA", fontsize=13, fontweight="bold")
ax[0].set_xlabel("Kategori CGPA")
ax[0].set_ylabel("Jumlah Mahasiswa")
ax[0].grid(axis="y", linestyle="--", alpha=0.5)
ax[0].legend()

# Nilai tertinggi
max_value = cgpa.max()
max_index = cgpa.idxmax()

ax[0].annotate(
    f"{max_value}",
    xy=(max_index, max_value),
    xytext=(0,15),
    textcoords="offset points",
    ha="center",
    fontsize=11,
    fontweight="bold",
    color="red"
)

# SUBPLOT 2 : Grafik Garis Academic Satisfaction
ax[1].plot(
    academic_labels,
    academic.values,
    marker="o",
    linewidth=2.5,
    color="darkorange",
    label="Academic Satisfaction"
)

ax[1].set_title("Distribusi Academic Satisfaction", fontsize=13, fontweight="bold")
ax[1].set_xlabel("Kategori Kepuasan")
ax[1].set_ylabel("Jumlah Mahasiswa")
ax[1].grid(True, linestyle="--", alpha=0.5)
ax[1].legend()

# Nilai tertinggi
max_value2 = academic.max()
max_index2 = academic.idxmax()

ax[1].annotate(
    f"{max_value2}",
    xy=(max_index2, max_value2),
    xytext=(0,15),
    textcoords="offset points",
    ha="center",
    fontsize=11,
    fontweight="bold",
    color="red"
)

# Memberikan ruang agar angka tidak bertabrakan
ax[0].set_ylim(0, max_value + 50)
ax[1].set_ylim(0, max_value2 + 50)

plt.suptitle(
    "Visualisasi Dataset Student Performance",
    fontsize=16,
    fontweight="bold"
)

plt.tight_layout()
plt.subplots_adjust(top=0.85)

# Simpan Grafik
plt.savefig("visualisasi_student_performance.png",
            dpi=300,
            bbox_inches="tight")

plt.savefig("visualisasi_student_performance.pdf",
            bbox_inches="tight")

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Membaca Dataset
df = pd.read_csv("student_performance_encoded.csv")

# Menampilkan Data
print("="*60)
print("10 BARIS PERTAMA DATASET")
print("="*60)
print(df.head(10))

print("\nJumlah Baris dan Kolom :", df.shape)

print("\nStatistik Deskriptif")
print(df.describe())

# Data Visualisasi
cgpa = df["cgpa"].value_counts().sort_index()
academic = df["Academic_Satisfaction"].value_counts().sort_index()

cgpa_labels = ["5.0–6.9", "7.0–8.4", "8.5–9.4", "9.5–10.0"]

academic_labels = [
    "Very\nUnsatisfied",
    "Unsatisfied",
    "Neutral",
    "Satisfied",
    "Very\nSatisfied"
]

# Membuat 2 Subplot
fig, ax = plt.subplots(1, 2, figsize=(17,7))

# SUBPLOT 1
bars = ax[0].bar(
    cgpa_labels,
    cgpa.values,
    color="skyblue",
    edgecolor="black",
    width=0.6,
    label="Jumlah Mahasiswa"
)

ax[0].set_title("Distribusi Kategori CGPA",
                fontsize=14,
                fontweight="bold")

ax[0].set_xlabel("Kategori CGPA")
ax[0].set_ylabel("Jumlah Mahasiswa")
ax[0].grid(axis="y", linestyle="--", alpha=0.5)
ax[0].legend()

max_value = cgpa.max()
max_index = cgpa.idxmax()

# Annotation
ax[0].annotate(
    f"Tertinggi\n{max_value}",
    xy=(max_index, max_value),
    xytext=(max_index, max_value + 55),
    ha="center",
    fontsize=10,
    color="red",
    fontweight="bold",
    arrowprops=dict(
        arrowstyle="->",
        color="red",
        lw=2
    )
)

ax[0].set_ylim(0, max_value + 90)

# SUBPLOT 2
ax[1].plot(
    academic_labels,
    academic.values,
    marker="o",
    markersize=8,
    linewidth=2.5,
    color="darkorange",
    label="Academic Satisfaction"
)

ax[1].set_title("Distribusi Academic Satisfaction",
                fontsize=14,
                fontweight="bold")

ax[1].set_xlabel("Kategori Kepuasan")
ax[1].set_ylabel("Jumlah Mahasiswa")
ax[1].grid(True, linestyle="--", alpha=0.5)
ax[1].legend()

max_value2 = academic.max()
max_index2 = academic.idxmax()

ax[1].annotate(
    f"Tertinggi\n{max_value2}",
    xy=(max_index2, max_value2),
    xytext=(max_index2, max_value2 + 55),
    ha="center",
    fontsize=10,
    color="red",
    fontweight="bold",
    arrowprops=dict(
        arrowstyle="->",
        color="red",
        lw=2
    )
)

ax[1].set_ylim(0, max_value2 + 90)

# Judul Utama
plt.suptitle(
    "Visualisasi Dataset Student Performance",
    fontsize=17,
    fontweight="bold"
)

plt.subplots_adjust(
    top=0.84,
    wspace=0.30
)

# Simpan Grafik
plt.savefig(
    "visualisasi_student_performance.png",
    dpi=300,
    bbox_inches="tight"
)

plt.savefig(
    "visualisasi_student_performance.pdf",
    bbox_inches="tight"
)

plt.show()

# Interpretasi

print("\nInterpretasi:")
print("- Mayoritas mahasiswa berada pada kategori CGPA tertentu sesuai batang tertinggi.")
print("- Tingkat Academic Satisfaction paling banyak ditunjukkan oleh titik tertinggi pada grafik.")
```
